1. Load the data

```python
import pandas as pd
WBO=pd.read_csv('//content/drive/MyDrive/Colab Notebooks/Outbreak Data-Table 1.csv')
WBO.head(10)
```

```python
most_populated_states = ['California', 'Texas', 'Florida']
```

```python
filtered_wbo_mps =WBO[WBO['State'].isin(most_populated_states)]
print(filtered_wbo_mps)
```

```python
mps_wbo = filtered_wbo_mps.dropna(axis=1)
print(mps_wbo)
```

```python
mps_wbo.sort_values(by='Month', inplace=True)
print(mps_wbo)
```

```python
column_names = mps_wbo.columns.tolist()
print("Column Names:")
for col in column_names:
  print(col)
```

```python
grouped_mps_wbo = mps_wbo.groupby('State')
```

```python
for state, group in grouped_mps_wbo:
  print(f"State: {state}")
  print(group)
  print('\n')
```
